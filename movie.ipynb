## Preprocessing
First, let's get all of the imports out of the way


```python
import pandas as pd
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

Now let's read each file into their Dataframes:


```python
movie_df=pd.read_csv('movies.csv')
reating_df=pd.read_csv('ratings_sample.csv')

movie_df.head()
```

Let's also remove the year from the **title** column by using pandas' replace function and store in a new **year** column.


```python
movie_df['year'] = movie_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
movie_df['year'] = movie_df.year.str.extract('(\d\d\d\d)',expand=False)
movie_df['title'] = movie_df.title.str.replace('(\(\d\d\d\d\))','',regex=True)
  
# movie_df.head()
 
movie_df.head()
 
```

```python
moviesWithGenres_df = movie_df.copy()

for index, row in movie_df.iterrows():
    for genre in row['genres'].split("|"):
         moviesWithGenres_df.at[index, genre] = 1
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
moviesWithGenres_df=moviesWithGenres_df.drop(['genres','year'],axis=1)
moviesWithGenres_df.head()
```

Let's start by extracting the genre table from the original dataframe:


```python
 

genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])

genreTable=genreTable.drop('movieId',axis=1)
genreTable=genreTable.drop('title',axis=1)



movie_rank=(np.array(genreTable*userProfile).sum(axis=1))/(np.array(userProfile).sum())

genreTable

data={'movieId':movie_df['movieId'],'rating':list(movie_rank)}

recommendationTable_df = pd.DataFrame(data, columns=['movieId', 'rating'])
recommendationTable_df
```

```python
recommendationTable_df=recommendationTable_df.sort_values(by='rating',ascending=False)
movieId_list=recommendationTable_df.head(20)['movieId']
movieId_list=list(movieId_list)
```
